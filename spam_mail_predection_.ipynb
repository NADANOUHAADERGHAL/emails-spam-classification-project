importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction .text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

data collection and preprossing


```python
# loading the data from csv file to a pandas Dataframe
email=pd.read_csv('/content/mail_data.csv')
email
```

```python
# replace the null values with a null string
mail_data = email.where((pd.notnull(email)),'')
```

```python
#print the first five rows
mail_data.head()
```

```python
#print the last five rows
mail_data.tail()
```

```python
#print the numbers of columns and rows
mail_data.shape
```

```python
#Label Encoding
#spam mail=0
#ham mail=1
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
mail_data
```

```python
# separating the data as texts and label

X = mail_data['Message']

Y = mail_data['Category']
```

```python
X
```

```python
Y
```


Splitting the data into training data & test data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
X_train.shape
```

```python
X_test.shape
```


featurs extraction



```python
# transform the text data to feature vectors that can be used as input to the Logistic regression
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
print(X_test_features)
```

training the model with the logistic regression algorithms

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with the training data
model.fit(X_train_features, Y_train)
```

evaluate and predecting the data


```python
# training the Logistic Regression model with the training data

prediction_training_data= model.predict(X_train_features)
accuracy_trainig_data=accuracy_score(Y_train, prediction_training_data)
```

```python
print(accuracy_trainig_data)
```

```python
# training the Logistic Regression model with the training data

prediction_testing_data= model.predict(X_test_features)
accuracy_testing_data=accuracy_score(Y_test, prediction_testing_data)
```

```python
print(accuracy_testing_data)
```

Building a Predictive System


```python

input_mail=["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's"]
# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
